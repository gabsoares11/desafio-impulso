# Desafio de Modelagem e Análise de Dados - ImpulsoGov

## Importando dados do <a href="http://datasource.coronacidades.org/help" style="color: gray">FarolCovid</a>

```python
import pandas as pd

farol_data = pd.read_csv("http://datasource.coronacidades.org/br/cities/cases/full")
farol_data.head()
```

## Importando dados do <a href="https://opendatasus.saude.gov.br/dataset/casos-nacionais" style="color: gray">OpenDataSUS</a>

```python
obitos_pb = pd.read_csv("data/obitos_pb.csv")
obitos_pb.head()
```

## Selecionando apenas os casos/óbitos ocorridos em Campina Grande

```python
farol_datacg = farol_data.loc[farol_data['city_name'] == 'Campina Grande']
farol_datacg.head()
```

```python
# Convertendo os objetos da coluna para DateTime
obitos_pb['Data do Óbito'] = pd.to_datetime(obitos_pb['Data do Óbito'])

obitos_cg = obitos_pb.copy()[obitos_pb['Município de Residência'] == 'Campina Grande']
obitos_cg.head()
```

## Visualização dos dados

```python
import numpy
import matplotlib.pyplot as plt

variables = obitos_cg['Sexo'].value_counts()
l = ['Masculino', 'Feminino']
c = ['#008B8B', '#00CED1']
explode = (0, 0.03)
fig1, ax1 = plt.subplots(figsize=(6,6))

plt.title('Visualização de óbitos por Covid-19 em Campina Grande por sexo', loc='center')
ax1.pie(variables, labels=l, colors=c, autopct='%0.2f%%', explode=explode);
```

```python
mortes_mes = obitos_cg.sort_values('Data do Óbito')
mortes_mes = mortes_mes['Data do Óbito'].groupby([mortes_mes['Data do Óbito'].dt.month, mortes_mes['Sexo']]).agg('count')
mortes_mes
```

```python
mortes_mes.groupby('Sexo').plot(figsize=(10,6));
plt.title('Visualização de mortes por mês classificado por sexo')
```
